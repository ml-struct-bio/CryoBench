```python
def natural_sort_key(s):
    # Convert the string to a list of text and numbers
    parts = re.split('([0-9]+)', s)
    
    # Convert numeric parts to integers for proper numeric comparison
    parts[1::2] = map(int, parts[1::2])
    
    return parts

def parse_class_assignments(path_for_label, path_for_model, K):
    # for labels
    cs = np.load(path_for_label)
    keys = ["alignments_class3D_{}/class_posterior".format(i) for i in range(K)]
    classes = [[x[k] for k in keys] for x in cs]
    classes = np.asarray(classes)
    class_id = classes.argmax(axis=1)
    
    # for models
    model_lst = []
    path_for_model.split('/')[-2]
    cs_model_path = [path_for_model+path_for_model.split('/')[-2]+"_passthrough_particles_class_{}.cs".format(i) for i in range(K)]
    for i in range(K):
        cs_model = np.load(cs_model_path[i])
        for j in range(len(cs_model)):
            num_vol = cs_model[j][1].split(b'_')[1].decode('utf-8')
            model_lst.append(num_vol)
    model_id = np.asarray(model_lst)
            
    return class_id, model_id
    
#     plt.savefig(f"{args.o}/{args.method}/{args.method}_{args.num_classes}.png")
```

```python
def plot_3dcls(axial_angles, dihedral_angles, labels_3dcls):
    x = np.repeat(axial_angles,1000)
    y = np.repeat(dihedral_angles,1000)
    xx = np.cos(x/180*np.pi)
    yy = np.sin(x/180*np.pi)
    xx += y/40
    jitter=0.04
    xx_jittered = xx + np.random.randn(len(xx))*jitter
    yy_jittered = yy + np.random.randn(len(yy))*jitter
    plt.xlabel("axial")
    plt.ylabel("dihedral")
    plt.scatter(xx_jittered,yy_jittered,cmap='tab10',s=1,alpha=.1,c=labels_3dcls, vmin=0,vmax=10)
```

```python
axial_angles = np.concatenate((np.arange(36)*5,-np.arange(1,37)*5,np.zeros(15)))
dihedral_angles = np.concatenate((np.zeros(72),np.arange(8)*5,-np.arange(1,8)*5))
```

# Use constant color scheme for all axial / dihedral + give a plot for the whole angles

```python
def plot_methods(v, axial_angles, dihedral_angles, is_umap=True):
    fig, axs = plt.subplots(1,3, figsize=(15,5))
    # Whole
    subset = v[:87000]
    c_all = np.repeat(axial_angles,1000)
    c = c_all
    axs[0].scatter(subset[:,0], subset[:,1], alpha=.1, s=1, cmap='gist_rainbow', c=c, vmin=-180, vmax=175)

#     fig, axs = plt.subplots(1,2, figsize=(10,5))
    plot_dim = (0,1)
    # Axial angles
    c_all = np.repeat(axial_angles,1000)
    c = c_all
    plot_args = dict(alpha=.1, s=1, cmap='gist_rainbow', vmin=-180,vmax=175)
    subset = v[:72000]
    c = c_all[:72000]
    axs[1].scatter(subset[:,plot_dim[0]], subset[:,plot_dim[1]], c=c, **plot_args)

    # Dihedral angles
    c_all = np.repeat(dihedral_angles,1000)
    c = c_all
    plot_args = dict(alpha=.1, s=1, cmap='gist_rainbow', vmin=-35,vmax=35)
    subset = v[72000:]
    c = c_all[72000:]
    axs[2].scatter(subset[:,plot_dim[0]], subset[:,plot_dim[1]], c=c, **plot_args)
    
    axs[0].set_title('Whole')
    axs[1].set_title('Axial')
    axs[2].set_title('Dihedral')
    axs[2].colorbar(label='Color Intensity')
    if is_umap:
        for ax in axs.flat:
            ax.set(xlabel='UMap1', ylabel='UMap2')
#         plt.savefig(f"{args.o}/{args.method}/{args.method}_umap.png")
    else:    
        for ax in axs.flat:
            ax.set(xlabel='Latent1', ylabel='Latent2')
#         plt.savefig(f"{args.o}/{args.method}/{args.method}_latent.png")
```

```python
antigens = []
vol_nums = []
for row in text_array:
    antigen = row.split(' ')[3].split('_')[0]
    vol_num = row.split(' ')[-1].split('_')[0]
    antigens.append(antigen)
    vol_nums.append(vol_num)
```

```python
val_lst = []
antigen_dict = {}
for i in range(len(vol_nums)):
    key = int(vol_nums[i])
    value = my_dict[antigens[i]]
    antigen_dict[key] = value
    val_lst.append(value)
```

```python
def natural_sort_key(s):
    # Convert the string to a list of text and numbers
    parts = re.split('([0-9]+)', s)
    
    # Convert numeric parts to integers for proper numeric comparison
    parts[1::2] = map(int, parts[1::2])
    
    return parts
```

```python
mix_sort_size = "/scratch/gpfs/ZHONGE/mj7341/research/cryobench/dataset/comp-het/mix-het/mix_order_size.txt"
mix_size = np.loadtxt(mix_sort_size, dtype='str')
```

```python
mix_current_name = '/scratch/gpfs/ZHONGE/mj7341/research/cryobench/dataset/comp-het/mix-het/add_noise/128/snr001/mrcs/sorted_particles.128.txt'
```

```python
mix_current_name = np.loadtxt(mix_current_name, dtype=str)
mix_current_name.shape
```

```python
lst = []
for i, org_name in enumerate(mix_current_name):
    for j, size_name in enumerate(mix_size):
        if size_name[1] in org_name:
            lst.append(size_name[0][:3])
```

```python
new_indices =np.array(lst).astype('int')*1000
```

```python
sequence = np.arange(1000)
reshaped_array = new_indices[:, np.newaxis]
expanded_array = reshaped_array + sequence
result_array = expanded_array.flatten()
new_umap = umap[result_array]
```

```python
np.arange(100)
```

```python
mix_label = np.arange(100)
fig, ax = plt.subplots(figsize=(4,4))
plot_dim = (0,1)
# Dihedral angles
c_all = np.repeat(mix_label,1000)
c = c_all
plot_args = dict(alpha=.1, s=1, cmap='gist_rainbow', vmin=0,vmax=100)
plt.scatter(new_umap[:,plot_dim[0]], new_umap[:,plot_dim[1]], c=c_all, **plot_args)
```
